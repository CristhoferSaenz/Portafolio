## <font color=#33A5FF>****3. Soluciones de preparación de datos****<tfont>

## 3.1. ¿Cómo convertir un tipo de dato num a cat?

Tipo de **dato incorrecto**: `estado_civil` (int64) 


### [solución]


### Convertimos solo una variable:

```python
# Hacemos una copia
data_original =data.copy()
```

```python
data["estado_civil_cat"]=data["estado_civil"].astype(str)
```

```python
# chequeo:
```

```python
status(data["estado_civil_cat"])
```

```python
data=data.drop("estado_civil",axis =1)
```

```python
status(data)
```

## 3.2. ¿Como crear la categoría 'Nulo'?

Hay `nan` en **categórica**: `nivel_educativo` 

### [solución]

### Utilizar método `fillna`

```python
status(data["nivel_educativo"])
```

```python
data["nivel_educativo"]=data["nivel_educativo"].fillna(value="nulo")
```

```python
status(data["nivel_educativo"])
```

```python
freq_tbl(data["nivel_educativo"])
```

## 3.3. ¿Cómo tratar los nulos en variables numéricas?


Hay `nan` en numérica: `edad`.



**Distintos enfoques**: 
  
  1. **Discretizar** (*num* a *cat*) y agregar el valor "**nulo**" (cómo paso anterior)


### [solución]]

### Discretización de edad

```python
data["edad_cat"],saved_bins=pd.qcut(data["edad"],q = 5, retbins = True)
```

```python
data["edad_cat"]
```

```python
freq_tbl(data["edad_cat"])
```

### Tiene 5 categorias y nulos

```python
#error
```

```python
#asignacion "nulo"
data["edad_cat"]=data["edad_cat"].cat.add_categories("nulo")
```

```python
data["edad_cat"]
```


### Ahora si podemos remplazar los nulos

```python
data["edad_cat"]=data["edad_cat"].fillna("nulo")
```

### Eliminamos la variable original

```python
data=data.drop(["edad"], axis=1)
```

```python
status(data)
```

```python
freq_tbl(data.edad_cat)
```

## <font color=#33A5FF>****4. Paso Final****<tfont>

## Todo a numérico (one hot encoding)

```python
data_final=pd.get_dummies(data,drop_first=True)
```

```python
data_final
```

```python
status(data_final)
```

### Guardamos los datos para usarlo para crear un modélo en "Machine Learning":

```python
import pickle
with open ("data/d_eph5.pickle","wb") as f:
    pickle.dump(data_final,f)
```

## 2. Creación de DataFrame

```python
#Librerías

import numpy as np
import pandas as pd
```

###  A partir de diccionarios de listas

```python
#Datos en listas

N = {"a": [10,15,20], "b": [12,17,14]}

#DataFrame

df_lista = pd.DataFrame(N)

#Impresión lista con print Spyder

print(df_lista)

#Impresión con función head()

df_lista.head()
```

```python
#Datos en listas

G = {"Nombres": ["Juan","Manuel","Teodoro","Mateo","Cesar"],
   "Promedio": [20,15,17,13,10],
    "Grupos": [1,2,1,2,3]}


#DataFrame

df_lista2 = pd.DataFrame(G)

#Impresión lista con print Spyder

print (df_lista2)

#Impresión con función head()

df_lista2.head()
```

```python
#Imprimir solo una columna Spyder

print(df_lista2["Promedio"])
```

```python
#Imprimir solo una columna

df_lista2["Promedio"]
```

##  Seleccionar un índice o columna

Una de las funciones que son muy fáciles en **Pandas** es seleccionar los datos que deseas en comparación con seleccionar un valor de una lista o un diccionario. Puedes seleccionar una columna de manera muy fácil, solamente deberás indicar el índice de la misma.

```python
#Seleccionar la primera columna del DataFrame

print("****La primera columna del DataFrame****")


#impresión

notas.iloc[:,0]
```

```python
#Seleccionar dos columnas del DataFrame

print("****Las dos columnas del DataFram****")

#impresión

notas.iloc[:, 0:2]
```

```python
#Seleccionar el valor de la primera fila y ultima columna del DataFrame

print("****El valor de la primera fila y ultima columna del DataFrame****")

#impresión

notas.iloc[0][2]
```

```python
#Seleccionar los valores de la primera fila del DataFrame

print("****Los valores de la primera fila del DataFrame****")

#impresión

notas.loc[0]
```

```python
#Seleccionar los valores de la primera fila del DataFrame

print("****Los valores de la primera fila del DataFrame****")

#impresión con iloc

notas.iloc[0,:]
```

### Operando entre columnas

```python
#Nueva columa a partir de Operación entre columnas

notas['Promedio 1 y 2'] = (notas['Nota1']+notas['Nota2'])/2

notas.head()
```

### Filtros
![filtros](https://user-images.githubusercontent.com/19308295/119282563-b6a12200-bbff-11eb-82ee-606fbfe2ac1b.png)

```python
#FILTRAR Cuantos alumnos califican como excelencia Promedio mayores o igual a 15
#MÉTODO 1

notas.loc[notas.Promedio >= 15]
```

```python
#FILTRAR Cuantos alumnos califican como excelencia Promedio mayores o igual a 12
#MÉTODO 2

notas.loc[notas['Promedio'] >= 12]
```
